```python
import mne
import pandas as pd
```

**Extract the first 150000 rows of data for each subject**

```python
# for subject 1-9
for i in range(1,10):
    file_path = f"derivatives/sub-00{i}/eeg/sub-00{i}_task-eyesclosed_eeg.set" 
    raw = mne.io.read_raw_eeglab(file_path, preload=True)
    data, times = raw.get_data(), raw.times 

    df = pd.DataFrame(data.T, columns=raw.ch_names)
    df *= 1e6
    df = df.iloc[:150000] # first 150000 rows of data
    df.to_csv(f"extracted/sub-00{i}_task-eyesclosed_eeg.csv", index=False)
```

```python
# for subject 10-88
for i in range(10,89):
    file_path = f"derivatives/sub-0{i}/eeg/sub-0{i}_task-eyesclosed_eeg.set" 
    raw = mne.io.read_raw_eeglab(file_path, preload=True)
    data, times = raw.get_data(), raw.times 

    df = pd.DataFrame(data.T, columns=raw.ch_names)
    df *= 1e6
    df = df.iloc[:150000] # first 150000 rows of data
    df.to_csv(f"extracted/sub-0{i}_task-eyesclosed_eeg.csv", index=False)
```
